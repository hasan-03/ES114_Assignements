# **PDF of CONTINUOUS RANDOM VARIABLE**

```python
import numpy as np
import matplotlib.pyplot as plt

def uniform_random(a,b):
  for i in range(len(a)):
    ai=a[i]
    bi=b[i]
    x_plot=np.linspace(-10,25,10000)
    y_plot=np.full_like(x_plot,0)
    for j in range(len(x_plot)):
      if x_plot[j]<bi and x_plot[j]>ai:
        y_plot[j]=1/(bi-ai)
    plt.plot(x_plot,y_plot,label=f'a={ai},b={bi}')
    plt.xlabel('x')
    plt.ylabel('Probability')
    plt.title('Uniform continuous distribution')
    plt.legend()
uniform_random([2,4,6],[4,8,9])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def gaussian(s,u):
  for i in range(len(s)):
    sigma=s[i]
    miu=u[i]
    x_plot=np.linspace(-10,10,1000)
    y_plot=np.full_like(x_plot,0)
    for j in range(len(x_plot)):
      y_plot[j]=(1/(np.sqrt(2*np.pi*sigma**2)))*np.exp(-((x_plot[j]-miu)**2)/(2*sigma**2))
    plt.plot(x_plot,y_plot,label=f'Sigma={sigma}, Miu={miu}')
    plt.xlabel('x')
    plt.ylabel('P(x)')
    plt.title('Gaussian Distribution')
    plt.legend()
    plt.fill(x_plot,y_plot)
gaussian([1,2,3],[0,-4,3])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def exponential(la):
  for i in range(len(la)):
    l=la[i]
    x_plot=np.linspace(0,10,1000)
    y_plot=np.full_like(x_plot,0)
    for j in range(len(x_plot)):
      y_plot[j]=l*np.exp(-l*x_plot[j])
    plt.plot(x_plot,y_plot,label=f'lambda={l}')
    plt.xlabel('x')
    plt.ylabel('P(x)')
    plt.title('Exponential Distribution')
    plt.legend()
    plt.grid()

exponential([1,0.5,1.5])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cauchy(a,b):
  for i in range(len(a)):
    Xm=a[i]
    Y=b[i]
    x_plot=np.linspace(-15,15,10000)
    y_plot=np.full_like(x_plot,0)
    for j in range(len(x_plot)):
      y_plot[j]=1/((np.pi*Y)*(1+((x_plot[j]-Xm)/Y)**2))
    plt.plot(x_plot,y_plot,label=f'Median={Xm}, Scale={Y}')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('P(x)')
    plt.title('Cauchy Distribution')
    plt.fill(x_plot,y_plot)
cauchy([0,2,-2],[0.5,1.5,1.5])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def laplacian(a,b):
  for i in range(len(a)):
    location=a[i]
    scale=b[i]
    x_plot=np.linspace(-10,10,1000)
    y_plot=np.full_like(x_plot,0)
    for j in range(len(x_plot)):
      y_plot[j]=(1/(2*scale))*np.exp(-np.absolute(x_plot[j]-location)/(scale))
    plt.plot(x_plot,y_plot,label=f'Location={location},Scale={scale}')
    plt.legend()

    plt.xlabel('x')
    plt.ylabel('P(x)')
    plt.title('Laplacian Distribution')
laplacian([0,5,-5],[2,3,3])
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rayleigh(s):
  for i in range(len(s)):
    sigma=s[i]
    x_plot=np.linspace(0,10,1000)
    y_plot=np.full_like(x_plot,0)
    for j in range(len(x_plot)):
      y_plot[j]=(x_plot[j]/(sigma**2))*np.exp((-x_plot[j]**2)/(2*sigma**2))
    plt.plot(x_plot,y_plot,label=f'Scale={sigma}')
    plt.legend()
    plt.title('Rayleigh')
    plt.xlabel('x')
    plt.ylabel('P(x)')
rayleigh([0.8,1,2,1.5])
plt.show()
```

# **CDF of CONTINUOUS RANDOM VARIABLE**

```python
import numpy as np
import matplotlib.pyplot as plt

def cdf_uniform_random(a,b):
  for i in range(len(a)):
    ai=a[i]
    bi=b[i]
    x_plot=np.linspace(0,15,10000)
    y_plot=np.full_like(x_plot,0)
    for j in range(len(x_plot)):
      if x_plot[j]<bi and x_plot[j]>ai:
        y_plot[j]=(x_plot[j]-ai)/(bi-ai)
      elif x_plot[j]>bi and x_plot[j]>ai:
        y_plot[j]=1
    plt.plot(x_plot,y_plot,label=f'a={ai},b={bi}')
    plt.xlabel('x')
    plt.ylabel('P(X<x)')
    plt.title('CDF Uniform continuous distribution')
    plt.legend()
plt.show()

cdf_uniform_random([2,4,6],[4,8,9])
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cdf_uniform_random(a,b):
  for i in range(len(a)):
    ai=a[i]
    bi=b[i]
    x_plot=np.linspace(0,15,10000)
    y_plot=np.full_like(x_plot,0)
    for j in range(len(x_plot)):
      if x_plot[j]<bi and x_plot[j]>ai:
        y_plot[j]=1/(bi-ai)
    y_plot=np.cumsum(y_plot)/np.sum(y_plot)
    plt.plot(x_plot,y_plot,label=f'a={ai},b={bi}')
    plt.xlabel('x')
    plt.ylabel('P(X<x)')
    plt.title('CDF with cumsum Uniform continuous distribution')
    plt.legend()
plt.show()

cdf_uniform_random([2,4,6],[4,8,9])
```

# **CDF of DISCRETE RANDOM VARIABLE**

```python
import numpy as np
import matplotlib.pyplot as plt

def cdf_dicrete_uniform(a,b):
  for i in range(len(a)):
    ai=a[i]
    bi=b[i]
    xplot=np.linspace(ai-2,2+bi,1000)
    yplot=np.full_like(xplot,0)
    for j in range(len(xplot)):
      if xplot[j]<=bi and xplot[j]>=ai:
        if int(xplot[j])==float(xplot[j]):
          yplot[j]=(xplot[j]-ai)/(bi-ai)
        else:
          yplot[j]=(int(xplot[j])-ai)/(bi-ai)
      elif xplot[j]>bi and xplot[j]>ai:
        yplot[j]=1
    plt.plot(xplot,yplot,label=f'a{i}={ai}, b{i}={bi}')
    plt.title('CDF of Discrete Uniform')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('F(x)')
cdf_dicrete_uniform([4,2,8],[9,5,15])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math as mt

def cdf_binomial(p,m):
  for i in range(len(p)):
    mi=m[i]
    pi=p[i]
    x_plot=np.arange(1,mi)
    y_plot=[]
    for j in x_plot:
      if j<mi:
        y_plot.append(mt.comb(mi,j)*((1-pi)**(mi-j))*(pi)**(j))
    yplot=[]
    for i in range(len(y_plot)-1):
      yplot+=[y_plot[i]]
      yplot+=[0]*100
    yplot+=[y_plot[len(y_plot)-1]]
    yplot=np.cumsum(yplot)/np.sum(yplot)
    xplot=np.linspace(1,mi,len(yplot))

    plt.title('CDF of Binomial Distribution')
    plt.xlabel('x')
    plt.ylabel('F(x)')
    plt.plot(xplot,yplot,label=f'p={pi},m={mi}')
    plt.legend()
cdf_binomial([0.5,0.8],[10,20])
```

```python
import matplotlib.pyplot as plt
import numpy as np

def cdf_bernoulli_plot(pmf):
  for i in range(len(pmf)):
    p=pmf[i]
    xplot=np.linspace(-2,2,1000)
    yplot=np.full_like(xplot,0)
    for j in range(len(xplot)):
      if xplot[j]>=0 and xplot[j]<1:
        yplot[j]=1-p
      elif xplot[j]>=1:
        yplot[j]=1
    plt.title('CDF of Bernoulli random Distribution')
    plt.plot(xplot,yplot,label=f'p={p}')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('F(x)')
    plt.show()
cdf_bernoulli_plot([0.3,0.8])
```

```python
import matplotlib.pyplot as plt
import numpy as np

def cdf_geometric(p,m):
  for i in range(len(p)):
    pi=p[i]
    mi=m[i]
    x_plot=np.arange(1,mi+1)
    y_plot=[]
    for j in range(len(x_plot)):
      y_plot.append((1-pi)**(x_plot[j]-1)*pi)
    yplot=[]
    for k in range(len(y_plot)):
      yplot+=[y_plot[k]]
      if k<len(y_plot)-1:
        yplot+=[0]*100
    yplot=np.cumsum(yplot)/np.sum(yplot)
    xplot=np.linspace(1,mi+1,len(yplot))
    plt.plot(xplot,yplot,label=f'p{i}={pi},m{i}={mi}')
    plt.title('CDF Geometric Distribution')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('F(x)')

cdf_geometric([0.2,0.5],[20,50])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math as mt

def cdf_poission(p,m):
    for i in range(len(p)):
      a=p[i]
      b=m[i]
      lb=a*b
      x_plot=np.arange(1,b+50)
      y_plot=[]
      for j in range(len(x_plot)):
        y_plot.append((np.exp(-lb)*lb**j)/mt.factorial(j))
      yplot=[]
      for i in range(len(y_plot)):
        yplot+=[y_plot[i]]
        if i<len(y_plot)-1:
          yplot+=[0]*1000
      xplot=np.linspace(1,b+50,len(yplot))
      yplot=np.cumsum(yplot)/np.sum(yplot)
      plt.plot(xplot,yplot,label=f'Lambda={lb}')
      plt.legend()
      plt.title('CDF Poission Distribution')
      plt.xlabel('x')
      plt.ylabel('F(x)')
cdf_poission([0.2,0.5],[50,100])
plt.show()
```
