##Question 1

Print the list of tournmaments in descending order in the year which had least number of matches in descending order.


```python
total_matches=np.zeros(25)
list_years=np.array([2000+i for i in range(25)])
j=0
for i in list_years:
  df=pd.read_csv(f'/content/drive/MyDrive/Probability_assignment10/atp_{i}.csv')
  total_matches[j]=df.shape[0]
  j+=1
dictionary={"Year":list_years,"Total Matches":total_matches}
data_matches=pd.DataFrame(dictionary)
display(data_matches)
plt.bar(list_years,total_matches)
plt.ylabel("Total Matches")
plt.xlabel("Year")
plt.title("Total Matches each Year")
plt.show()
```

##Question 2

Find the variance of Winning player rank and Lossing player rank find the correlation coefficient?

```python
loser_ranks=[]
winner_ranks=[]

for i in list_years:
  df=pd.read_csv(f'/content/drive/MyDrive/Probability_assignment10/atp_{i}.csv')
  df1=df['WRank']
  winner_ranks.append(df1)
data_winner_ranks=pd.concat(winner_ranks,ignore_index=True)
df2=data_winner_ranks.str.contains('NR',na=False)
filtered_winner_ranks=data_winner_ranks[~df2]

numeric_winner_ranks=pd.to_numeric(filtered_winner_ranks)

for i in list_years:
  df=pd.read_csv(f'/content/drive/MyDrive/Probability_assignment10/atp_{i}.csv')
  df1=df['LRank']
  loser_ranks.append(df1)
data_loser_ranks=pd.concat(loser_ranks,ignore_index=True)
df2=data_loser_ranks.str.contains('NR',na=False)
filtered_loser_ranks=data_loser_ranks[~df2]

numeric_loser_ranks=pd.to_numeric(filtered_loser_ranks)

print("Correlation Corfficient Between Winner ranks and Loser Ranks",numeric_winner_ranks.corr(numeric_loser_ranks))
```

##Question 4

Print the name of best and worst player for each surface.


```python
data=[]

for i in list_years:
  df=pd.read_csv(f'/content/drive/MyDrive/Probability_assignment10/atp_{i}.csv')
  data.append(df)

combined_data=pd.concat(data)

combined_data=combined_data.dropna(subset=['Surface','Winner','Loser'])
matches_won = combined_data.groupby(['Surface', 'Winner']).size().reset_index(name='Winned')
matches_lost = combined_data.groupby(['Surface', 'Loser']).size().reset_index(name='Lost')

total_matches = pd.merge(matches_won, matches_lost, left_on=['Surface', 'Winner'], right_on=['Surface', 'Loser'], how='outer')
total_matches['Winned'] = total_matches['Winned'].fillna(0)
total_matches['Lost'] = total_matches['Lost'].fillna(0)

total_matches['TotalMatches'] = total_matches['Winned'] + total_matches['Lost']

total_matches['WinRate'] = total_matches['Winned'] / total_matches['TotalMatches']
total_matches['loseRate'] = total_matches['Lost'] / total_matches['TotalMatches']

best_players = total_matches.loc[total_matches.groupby('Surface')['WinRate'].idxmax()]
print("Best Player for each Surface")
data_best=pd.DataFrame(best_players[['Surface', 'Winner', 'WinRate']])
display(data_best)

worst_players = total_matches.loc[total_matches.groupby('Surface')['loseRate'].idxmax()]
print("Worst Player for each Surface")
data_worst=pd.DataFrame(worst_players[['Surface', 'Loser', 'loseRate']])
display(data_worst)
```

##Question 5
Plot the gaussian curve of BMI of all the winner palyers.

```python
merged_data_winner=pd.merge(combined_data,atp_players,left_on='Winner',right_on='name',how='right')

merged_data_winner['BMI']=(merged_data_winner['weight_kg'])/(((merged_data_winner['height_cm'])/100)**2)

bmi = merged_data_winner['BMI']
bmi = bmi[(bmi >= 10) & (bmi <= 50) & ~bmi.isna()]
mean,std = norm.fit(bmi)

plt.hist(bmi, bins=30, density=True, alpha=0.5, color='b')

xmin,xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std)
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel('BMI')
plt.ylabel('Density')
plt.title('Gaussian (Normal) Distribution of BMI')
plt.show()
```
