**Joint Probability**:

PXY(A,B)=P(X=A,Y=B) - for the case of 2 random variables taking the values A and B resp.

The probability of 2 or more events occuring. It combines the marginal and conditional Probability distributions. In case of independent random variables, the Joint PMF is given by:

P(A and B) = P(A) * P(B)

In the first question, this is what needs to be verified. Here, P(A) and P(B) are the marginal PMFs. The summation of the each row and column would give the corresponding marginal PMFs. In the example given below, the sum of all the probabilities in the 1st column (-10.0) would give the marginal probability of -10.0 i.e, P(Y=-10.0).

Note:
Given marginal PMFs, we can get joint PMF only when A and B are independent. But, given the joint PMF, we can always get the marginal PMFs.

The below is a sample joint PMF (randomly generated).

```python
import numpy as np

# Generating a matrix of random values for Joint pmf
def random_joint_pmf(x_size, y_size):
    random_mat = np.random.rand(x_size, y_size)
    joint_pmf = random_mat / random_mat.sum()
    return joint_pmf

# creating the joint PMF table (Just for visualising the table)
def joint_pmf_table(joint_pmf, x_values, y_values):
    print("    ", end="")
    for y in y_values:
        print(f"{y:>10}", end="")
    print()
    for i, row in enumerate(joint_pmf):
        print(f"{x_values[i]:<4}", end="")
        for prob in row:
            print(f"{prob:>10.4f}", end="")
        print()
```

```python
# x_values = ['x1', 'x2', 'x3']
# y_values = ['y1', 'y2', 'y3', 'y4']

x_values = np.linspace(10,20,5)
y_values = np.linspace(-10,-1,5)
joint_pmf = random_joint_pmf(len(x_values), len(y_values))
joint_pmf_table(joint_pmf, x_values, y_values)
```

**Plotting Joint and marginal PMFs and CDFs**

this would require us to use 3D plots. Take the X and Y var values in two of the axes and probabilty values in the 3rd axis.

```python
#examples of 3D discrete and continuous plots using matplotlib
'''
Code taken from : https://matplotlib.org/stable/gallery/mplot3d/3d_bars.html
'''
import matplotlib.pyplot as plt
import numpy as np

# 3D Discrete Plot for plotting PMFs
# set up the figure and axes
fig = plt.figure(figsize=(8, 3))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')

# fake data
_x = np.arange(4)
_y = np.arange(5)
_xx, _yy = np.meshgrid(_x, _y)
x, y = _xx.ravel(), _yy.ravel()

top = x + y
bottom = np.zeros_like(top)
width = depth = 1

ax1.bar3d(x, y, bottom, width, depth, top, shade=True)
'''
x, y, bottom: array-like
The coordinates of the anchor point of the bars.
 width, depth, top: array-like
The width, depth, and height of the bars, respectively.
'''
ax1.set_title('Shaded')

ax2.bar3d(x, y, bottom, width, depth, top, shade=False)
ax2.set_title('Not Shaded')

plt.show()
```

```python
# COntinuous 3D plot (surface plot) for plotting CDFs

import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
from matplotlib.ticker import LinearLocator

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)

X, Y = np.meshgrid(X, Y)
'''
The np.meshgrid function is used to create a rectangular grid out of two given 1D arrays
'''
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

'''
where X and Y are 2D array of points of x and y while Z is 2D array of heights.
'''

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
'''
We can add colormap to show which regions correspond to what values. It basically provides a mapping from scalar values to colors.
This is done for better data interpretation
'''
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
