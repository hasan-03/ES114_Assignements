## **Discrete Uniform**

```python
import numpy as np
import matplotlib.pyplot as plt

def dicrete_uniform(Events):
  for i in range(len(Events)):
    n=Events[i]                    #for Discrete uniform probability distribution we will append all the value of events that are from -n to n in x axis using for loop and append
    xplot=[]                       #the probability of each event is uniform in discrete uniform and value is plotted in y axis which is equal to 1/(total number of events(2n+1))
    yplot=[]
    for i in range(-n,n+1):
      xplot.append(i)
    print(xplot)
    for j in range(2*n+1):
      yplot.append(1/(2*(n)+1))
    print(yplot)
    plt.bar(xplot,yplot,width=0.2,color='k')
    plt.title(f'Total Events={2*n+1}')
    plt.xlabel('Event')
    plt.ylabel('Probability')
    plt.grid()
    plt.show()
dicrete_uniform([5,10,20])
```

# **Binomial Distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
def factorial(n):
  if n==0 or n==1:
    return 1
  else:
    return n*factorial(n-1)                     #define factorial of number n

def comb(a,b):
  if a<b:
    return 'invalide a,b'                        #define combination of a,b where a>b
  else:
    return factorial(a)/(factorial(b)*factorial(a-b))
def binomial_plot(p,m):
  for i in range(len(p)):
    xplot=[]                             #in the binomial plot we are plotting values of m on x axis and the values of probability on y axis
    yplot=[]
    a=p[i]
    b=m[i]
    for j in range(b+1):
      xplot.append(j)
      yplot.append(comb(b,j)*((1-a)**(b-j))*(a)**(j))
    print(xplot)
    print(yplot)
    plt.bar(xplot,yplot,width=0.3)
    plt.plot(xplot,yplot,color='b',marker="o")
    plt.title(f'p={a},m={b}')
    plt.xlabel('Event')
    plt.ylabel('Probability')
    plt.grid()
    plt.show()
binomial_plot([0.2,0.4,0.8],[30,50,100])
```

# **Geometric Distribution**

```python
import numpy as np
import matplotlib.pyplot as plt

def geometric_plot(p,m):
  for i in range(len(p)):
    a=p[i]
    b=m[i]
    xplot=[]
    yplot=[]
    for j in range(1,b+1):
      x=j
      xplot.append(x)
      y=a*(1-a)**(j-1)
      yplot.append(y)
    print(xplot)
    print(yplot)
    plt.bar(xplot,yplot,width=0.3)
    plt.plot(xplot,yplot)
    plt.title(f'p={a}, m={b}')
    plt.xlabel('Event')
    plt.ylabel('Probability')
    plt.grid()
    plt.show()
geometric_plot([0.2,0.5,0.8],[20,30,40])
```

# **Poission Distribution**

```python
import matplotlib.pyplot as plt
import numpy as np

def poission_plot(p,m):
  for i in range(len(p)):
    a=p[i]
    b=m[i]
    lb=a*b
    x=[]
    for j in range(b+1):
      x.append(j)
    y=[]
    for k in x:
      y.append(((np.exp(-lb))*(lb**k))/np.math.factorial(k))
    print(x)
    print(y)
    plt.bar(x,y,color='b',width=0.3)
    plt.plot(x,y,marker='*')
    plt.title(f'p={a},m={b}')
    plt.xlabel('Event')
    plt.ylabel('Probability')
    plt.grid()
    plt.show()
poission_plot([0.2,0.4,0.8],[30,50,100])

```
